```python
def extract_decks_links(soup):
    list_of_decks = []
    try:
        decks_links = soup.find('div',attrs= {'class': 'elementor-posts-container elementor-posts elementor-posts--skin-cards elementor-grid'}).find_all('a')
        for link in decks_links:
            deck_link = link.get('href')
            list_of_decks.append(deck_link)
        return list_of_decks
    except Exception as e:
        print('Error whit that link/name')
        print(e)
        
```

```python
decks_links = extract_decks_links(BeautifulSoup(requests.get('https://hearthstone-decks.net/standard-decks/druid//').text, 'lxml'))
```

```python
def extract_info_deck(soup):
    deck = {}
    try:
        deck_name = soup.find('div', attrs= {'class': 'entry-title'})
        if deck_name:
            deck['deck_name'] = deck_name.find('h1').text
        else:
            deck['deck_name'] = None
        
        deck_code = soup.find('input', attrs= {'id': 'Code1'})
        if deck_code:
            deck['deck_code'] = deck_code.get('value')
        else:
            deck['deck_code'] = None
            
        deck_date = soup.find('ul', attrs= {'class': 'entry-meta list-inline'})
        if deck_date:
            deck['deck_date'] = deck_date.find('li').text
        else:
            deck['deck_date'] = None
        
        deck_tags = soup.find('li', attrs= {'class': 'meta-category'})
        if deck_tags:
            deck['deck_tags'] = deck_tags.text
        else:
            deck['deck_tags'] = None
        
        return deck
    except Exception as e:
        print('Error')
        print(e)
```

```python
extract_info_deck(BeautifulSoup(requests.get('https://hearthstone-decks.net/prince-renathal-kazakusan-druid-96-legend-ttvdoodle-score-7-3/').text, 'lxml'))
```

```python
def scrape(url):
    try:
        deck = requests.get(url)
    except Exception as e:
        print('Error whit', url)
        print(e)
        return None
    
    if deck.status_code != 200:
        print(f'Error scraping this deck {url}')
        print(f'Error: {url.status_code}')
        return None
    
    soup_deck = BeautifulSoup(deck.text, 'lxml')
    deck = extract_info_deck(soup_deck)
    deck['url'] = url
    return deck
```

```python
print(scrape('https://hearthstone-decks.net/prince-renathal-kazakusan-druid-96-legend-ttvdoodle-score-7-3/'))
```

```python
decks_data = []
for i,deck in enumerate(decks_links):
    print(f'Deck{i+1}/{len(decks_links)}')
    decks_data.append(scrape(deck))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(decks_data)
df
```
