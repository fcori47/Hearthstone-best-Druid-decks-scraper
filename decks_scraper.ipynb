```python
import requests
from bs4 import BeautifulSoup
import pymysql
```

```python
dict_classes_links = {'Demon Hunter': 'https://hearthstone-decks.net/standard-decks/demon-hunter/','Druid':'https://hearthstone-decks.net/standard-decks/druid/','Hunter':'https://hearthstone-decks.net/standard-decks/hunter/','Mage':'https://hearthstone-decks.net/standard-decks/mage/','Paladin': 'https://hearthstone-decks.net/standard-decks/paladin/', 'Priest': 'https://hearthstone-decks.net/standard-decks/priest/', 'Rogue': 'https://hearthstone-decks.net/standard-decks/rogue/', 'Shaman': 'https://hearthstone-decks.net/standard-decks/shaman/', 'Warlock': 'https://hearthstone-decks.net/standard-decks/warlock/', 'Warrior': 'https://hearthstone-decks.net/standard-decks/warrior/'}
```

```python
def extract_class_link(hs_class:str):
    try:
        if hs_class == 'Demon Hunter' or 'DH': url = dict_classes_links['Demon Hunter']
        if hs_class == 'Druid': url = dict_classes_links['Druid']
        if hs_class == 'Hunter': url = dict_classes_links['Hunter']
        if hs_class == 'Mage': url = dict_classes_links['Mage']
        if hs_class == 'Paladin': url = dict_classes_links['Paladin']
        if hs_class == 'Priest': url = dict_classes_links['Priest']
        if hs_class == 'Rogue': url = dict_classes_links['Rogue']
        if hs_class == 'Shaman': url = dict_classes_links['Shaman']
        if hs_class == 'Warlock': url = dict_classes_links['Warlock']
        if hs_class == 'Warrior': url = dict_classes_links['Warrior']
        return url
    except Exception as e:
        print(f'Error: {e}')
```

```python
def extract_deck_links(soup):
    list_of_decks = []
    try:
        decks_links = soup.find('div',attrs= {'class': 'elementor-posts-container elementor-posts elementor-posts--skin-cards elementor-grid'}).find_all('a')
        for link in decks_links:
            deck_link = link.get('href')
            list_of_decks.append(deck_link)
        return list_of_decks
    except Exception as e:
        print('Error whit that link/name')
        print(e)
        
```

```python
decks_link = extract_deck_links(BeautifulSoup(requests.get(extract_class_link('Hunter')).text, 'lxml'))
```

```python
def extract_info_deck(soup):
    deck = {}
    try:
        deck_name = soup.find('div', attrs= {'class': 'entry-title'})
        if deck_name:
            deck['deck_name'] = deck_name.find('h1').text
        else:
            deck['deck_name'] = None
        
        deck_code = soup.find('input', attrs= {'id': 'Code1'})
        if deck_code:
            deck['deck_code'] = deck_code.get('value')
        else:
            deck['deck_code'] = None
            
        deck_date = soup.find('ul', attrs= {'class': 'entry-meta list-inline'})
        if deck_date:
            deck['deck_date'] = deck_date.find('li').text
        else:
            deck['deck_date'] = None
        
        deck_tags = soup.find('li', attrs= {'class': 'meta-category'})
        if deck_tags:
            deck['deck_tags'] = deck_tags.text
        else:
            deck['deck_tags'] = None
        
        return deck
    except Exception as e:
        print('Error')
        print(e)
```

```python
def scrape(url):
    try:
        deck = requests.get(url)
    except Exception as e:
        print('Error whit', url)
        print(e)
        return None
    
    if deck.status_code != 200:
        print(f'Error scraping this deck {url}')
        print(f'Error: {url.status_code}')
        return None
    
    soup_deck = BeautifulSoup(deck.text, 'lxml')
    deck = extract_info_deck(soup_deck)
    deck['url'] = url
    return deck
```

```python
decks_data = []
for i,deck in enumerate(decks_link):
    print(f'Deck{i+1}/{len(decks_link)}')
    decks_data.append(scrape(deck))
```

```python
counter = 0
while counter < 1:
    selection = input(f'Elija una clase de Hearthstone de la cuál conseguir sus mejores mazos, Las opciones son: {[i for i in dict_classes_links.keys()]})')
    if selection == 'Demon Hunter' or 'DH': hs_class = extract_class_link('Demon Hunter')
    if selection == 'Druid': hs_class = extract_class_link('Druid')
    if selection == 'Hunter': hs_class = extract_class_link('Hunter')
    if selection == 'Mage': hs_class = extract_class_link('Mage')
    if selection == 'Paladin': hs_class = extract_class_link('Paladin')
    if selection == 'Priest': hs_class = extract_class_link('Priest')
    if selection == 'Rogue': hs_class = extract_class_link('Rogue')
    if selection == 'Shaman': hs_class = extract_class_link('Shaman')
    if selection == 'Warlock': hs_class = extract_class_link('Warlock')
    if selection == 'Warrior': hs_class = extract_class_link('Warrior')
    
    hs_class_decks_link = extract_deck_links(BeautifulSoup(requests.get(hs_class).text, 'lxml'))
    
    hs_class_decks_data = []
    for i,deck in enumerate(hs_class_decks_link):
        print(f'Deck{i+1}/{len(hs_class_decks_link)}')
        hs_class_decks_data.append(scrape(deck))
    
    print('Scraping finalizado.')
    break
        
```
